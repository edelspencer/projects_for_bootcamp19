This dataset contains descriptive details of residential properties on the real estate market in Washington, DC, in 2018. They were built from 1947 to the present time. It lists the features in 49 columns. There is some missing data. This is a high price area - mean price is $931351.59. They are either semi-detached or row houses. Condition varies from 'fair' to 'very good'. Each one has a range of 1-4 condominium units. There are no condominiums on the market in zip code 20009.

Question: Identify predictive features for the price of a house.  What attributes are most relevant?

```python
#Import dataset DC_Properties.csv

import pandas as pd
import numpy as np

location = "datasets/DC_Properties.csv"
df = pd.read_csv(location)
```

```python
#Create a new dataframe with only the data that I want to consider 
df2 = df[["PRICE","ROOMS","BEDRM","BATHRM","HF_BATHRM","KITCHENS","NUM_UNITS","SALE_NUM","GBA","AYB","ZIPCODE","LATITUDE","LONGITUDE"]]
df2.head()
```

```python
#Create a correlation table to determine which variables to include in regression 
df2.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
#Run a regression on above variables, except half-bathrooms and rooms
result = smf.ols('PRICE ~  SALE_NUM + GBA + AYB + ZIPCODE + LATITUDE + LONGITUDE + NUM_UNITS + KITCHENS + BATHRM + BEDRM' , data=df).fit()
```

```python
result.summary()
```

```python
#Remove the intercept - set coefficient to 0
result = smf.ols(formula='PRICE ~ SALE_NUM + GBA + AYB + ZIPCODE + LATITUDE + LONGITUDE + NUM_UNITS + KITCHENS + BATHRM + BEDRM- 1', data=df).fit()
result.summary()
```

From the above, we can deduce that the following features predict the price of properties:  number of sales, gross building area, actual year built, zip code, location (latitude + longitude), number of units in building, number of kitchens, bathrooms and bedrooms. 
The most relevant attributes are: zip code and actual year built.
