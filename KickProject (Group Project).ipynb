```python
# Let's visualize the different project status for the us_pledged_real

sns.set()
sns.barplot('state','usd_pledged_real',data=df)
```

```python
df5 = df.loc[df['state']=='successful']
df5.head()
```

```python
# Analyze the successful project to create a model. Transforming the successful state attributes using dummies

df6 = pd.get_dummies(data=df5, columns=['main_category'])
df6.head()
```

```python
# Summarizing the data using pivot table
df7 = pd.pivot_table(df6,index=('category'),values =['backers','usd_pledged_real','usd_goal_real'],aggfunc='sum')
df7
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df7)
```
