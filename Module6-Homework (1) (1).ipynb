```python
import statsmodels.formula.api as smf
```

```python
#Run a correlation on the dataset.
df.corr()
```

```python
#1. Run a regression.   How does price relate to carat, x, y and z? 
result = smf.ols('price ~ carat + x + y + z', data=df).fit()
```

```python
result.summary()
```

Result: adj.R-squared is high (0.854) so this model fits; intercept, carat, x and y are statistically significant variables; p-value (0.000) says this is a statistically significant result.  
    

```python
#Remove y from correlation - not strong relationship
result = smf.ols('price ~ carat + x + z', data=df).fit()
result.summary()
```

Result: Adj.R-squared remains the same; intercept increases while x and z decrease; p-value is same. Not much change. Something else is influencing the model.

```python
#set coefficient to 0
result = smf.ols(formula='price ~ carat + x + y + z - 1', data=df).fit()
result.summary()
```

Result: Adj.r-squared increases(0.926); p-value remains the same; carat is the only IV that increases. 
I conclude that it is the only factor that influences price.   

```python
#2.  How does price relate to carat, depth and table? 
result = smf.ols('price ~ carat + depth + table', data=df).fit()
```

```python
result.summary()
```

Result: adj.R-squared is high (0.854) so this model fits; intercept doesn't make sense; p-value is optimal (0.000); carat is far removed from depth and table.   

```python
#remove carat from regression
result = smf.ols('price ~ depth + table', data=df).fit()
result.summary()
```

Results: Adj.R-square plunged so this is not a good model.  Let's try with different IVs. 

```python
#remove depth from regression
result = smf.ols('price ~ carat + table', data=df).fit()
result.summary()
```

Results:  Adj.R-square is back up but not as high as the first model; p-value is same; table is lower. 
    Will use the first model in next step.    

```python
#set coefficient to 0
result = smf.ols(formula='price ~ carat + depth + table - 1', data=df).fit()
result.summary()
```

Results: Adj-R-squared increased so model is a good fit; p-value is same; carat is the only IV that increased.
Conclusion - only carat size influences price.  

```python
Location = "datasets/tamiami.csv"

df = pd.read_csv(Location)
df.head()
```

```python
columns = ['location', 'sales', 'employees', 'restaurants', 'foodcarts', 'price']

#change column names for readability
df.columns = columns
df.head()
```

```python
#Number of records in dataset
df.tail()
```

```python
#show data types in the dataset
df.dtypes
```

```python
#Clean the data:
#Create list of true/false values
dupe = df.duplicated() 
#Shwo rows where duplicate is true
df[dupe] 
```

```python
#drop rows with any missing data
df_no_missing = df.dropna()
df_no_missing
```

```python
#Run a correlation on the dataset.
df.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
#Regression
result = smf.ols('sales ~ employees + restaurants + foodcarts + price', data=df).fit()
```

```python
result.summary()
```

Results: Adj.R-squared is high (0.977) so model fits; intercept, foodcarts and price are significant IVs; p-factor significant. 

```python
#Remove restaurants
result = smf.ols('sales ~ employees + foodcarts + price', data=df).fit()
result.summary()
```

```python
Adj-R-squared remains the same; p-factors significant.
```

```python
#Run all variables without the intercept - set coefficient to 0
result = smf.ols(formula='sales ~ employees + restaurants + foodcarts + price - 1', data=df).fit()
result.summary()
```

```python
Adj-R-squared improved (0.986); p-factor for price is high (0.445).
```

```python
#Run above model without price 
result = smf.ols(formula='sales ~ employees + restaurants + foodcarts  - 1', data=df).fit()
result.summary()
```

```python
Adj.R-squared remains the same; p-factors improve. 
```
