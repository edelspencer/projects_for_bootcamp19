```python
Location = "datasets/tamiami.csv"

df = pd.read_csv(Location)
df.head()
```

```python
columns = ['location', 'sales', 'employees', 'restaurants', 'foodcarts', 'price']

#change column names for readability
df.columns = columns
df.head()
```

```python
#Number of records in dataset
df.tail()
```

```python
#show data types in the dataset
df.dtypes
```

```python
#Clean the data:
#Create list of true/false values
dupe = df.duplicated() 
#Shwo rows where duplicate is true
df[dupe] 
```

```python
#drop rows with any missing data
df_no_missing = df.dropna()
df_no_missing
```

```python
#Run a correlation on the dataset.
df.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
#Regression
result = smf.ols('sales ~ location + employees + price', data=df).fit()
```

```python
result.summary()
```

```python
#remove price from regression
result = smf.ols('sales ~ location + employees', data=df).fit()
result.summary()
```

```python
#set coefficient to 0
result = smf.ols(formula='sales ~ location + employees + price - 1', data=df).fit()
result.summary()
```

Takeaways:  The most important variable in sales volume is location.
