```python
#Import dataset DC_Properties.csv

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline

location = "datasets/DC_Properties.csv"
df = pd.read_csv(location)
```

```python
#Lets see what the database looks like
df.head()
```

```python
#show the column names in the dataset
df.columns
#or print(df.columns)
```

```python
#show data types in the dataset
df.dtypes
```

```python
#show number of records in dataset
df.tail()
```

```python
#print dataset with zipcode and price columns 
pd.pivot_table(df, values=['ZIPCODE'], index=['PRICE'])
```

```python
#make a scatterplot
sns.lmplot(x='PRICE', y='ZIPCODE', data=df)
```

```python
#Boxplot of above 
sns.boxplot(data=df)
```

```python
#Boxplot was unintelligible so change size of plot
plt.subplots(figsize=(13,7))

#violin plot shows the distribution of number of units by zipcode
sns.violinplot(x='ZIPCODE', y='NUM_UNITS', data=df)
```

```python
#set index as zip code
df.set_index('ZIPCODE', inplace=True)
df.head()
```

```python
#make a scatterplot
sns.lmplot(x='PRICE', y='LANDAREA', data=df)
```

```python
#use two columns to groupby
df.groupby(['ZIPCODE', 'PRICE']).mean()
```

```python
#pivot table default function is mean
pd.pivot_table(df, index=['ZIPCODE'])
```

```python
#create a dataframe containing the room description for listings by zip code

df2 = df[['ZIPCODE','BATHRM','HF_BATHRM','BEDRM','ROOMS']]
df2.head()
```

```python
#create color palette
#color Hex numbers are from Bulbapedia (https://bulbapedia.bulbagarden.net/wiki/Category:Type_color_templates)
roomsbyzip_colors = ['#78C850',  # BATHRM
                    '#F08030',  # HF_BATHRM
                    '#6890F0',  # ROOMS
                    '#A8B820',  # BEDROOM
                    '#A8A878',  # FIREPLACES
                    '#A040A0',  # ZIPCODE
                    '#F8D030',  # PRICE
                                       ]
```

```python
#create a bar plot of room availability
sns.countplot(x='ROOMS',
              data=df, 
              palette=roomsbyzip_colors)

plt.xticks(rotation=-45)
```
